```python
import os

# 1. Zabardasti '.streamlit' naam ka hidden folder banana
os.makedirs(".streamlit", exist_ok=True)

# 2. Config file ka code
config_code = """
[theme]
base = "dark"
primaryColor = "#3b82f6"
backgroundColor = "#0f172a"
secondaryBackgroundColor = "#1e293b"
textColor = "#f8fafc"
font = "sans serif"

[client]
toolbarMode = "minimal"
"""

# 3. Folder ke andar 'config.toml' file banakar code save karna
with open(".streamlit/config.toml", "w") as f:
    f.write(config_code.strip())

print("✅ Success! .streamlit folder aur config.toml file successfully ban gayi hai!")
```
